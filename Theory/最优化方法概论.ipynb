### 最优化
最优化是在工程领域非常重要的一门学科，几乎所有的问题到最后都会化成一个求极值的问题，因此掌握必备的最优化方法是非常有必要的。
### 学习方法
知道每个优化算法工作是什么样的，翻译成能解释的语言，用途在哪以及优缺点。
### 解决步骤
1.根据实际模型转化为最优化问题:  
    1.1 优化变量  
    1.2 优化目标  
    1.3 优化约束  
2.选择合适的方法来进行极值求解。  
最优化方法的定义：对于单个或者多个变量x，在满足一定条件下，使得f（x）值最小。

### 优化问题分类
#### 根据约束
1、无约束优化问题  
2、约束优化问题  
2.1、等式约束问题  
2.2、不等式约束问题  
2.3、混合约束问题  
#### 根据目标函数和约束函数类型分类
1、目标函数和不等式约束条件均为线性称为线性规划  
2、不全为线性的时候称为非线性优化  
3、当不等式约束为线性，目标函数为为二次的称为二次规划   
ps: 一般来说, 我们都可以将目标函数及约束用数学公式表达出来 (这就是建模), 但对于某些特别的"黑箱"问题, 可能我们并不知道它们的具体形式, 一般会假设, 给定一个输入, 会产生一个输出. 无导数优化方法 (derivative-free optimization, DFO) 就是一类用来解决此类问题的算法.

### 最优解最优值
全局最优解x,全局最优值f（x）  
局部最优解x,局部最优解f（x）  

### 凸优化
凸优化是一类特殊的优化方法，可以分为两类：  
1、无约束凸优化问题：无约束且目标函数为凸函数的优化问题，局部最优解即为全局最优解。  
2、有约束凸优化问题：目标函数和不等式函数均为凸函数且等式约束为仿射函数，利用拉格朗日求解就是最优解。

### 常见方法
1、梯度下降法（最速下降法）  
2、牛顿法  
3、拟牛顿法  
4、共轭梯度法  
5、拉格朗日约束优化  
6、启发式算法
