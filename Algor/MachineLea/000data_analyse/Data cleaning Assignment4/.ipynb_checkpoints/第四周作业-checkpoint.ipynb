```python
gdp = pd.read_excel(r'C:\Users\MrFive1001\Documents\PycharmProjects\Kaggle\Data cleaning Assignment4\gdplev.xls',skiprows=range(5))
#得到一个下列数据
gdp
```

```python
#我们需要抛弃第0，1行，第4，7列
# gdp = gdp.iloc[list((range(len(gdp.index))) - [0,1]),(list(range(len(gdp.columns)))-[3,4,7])] 不能使用
gdp = ((gdp.drop(gdp.index[0:2],axis = 0)).drop(gdp.columns[3:4].append(gdp.columns[7:8]),axis = 1)) #注意注意index连接方式append
gdp
```

```python
# 我们再把gdp分为两部分
gdp_year = gdp.iloc[:,0:3].dropna(how = 'all').copy()
gdp_year.index = range(len(gdp_year.index))
gdp_year.columns = ['year','GDP1','GDP2']
gdp_year
```

```python
gdp_quarter = gdp.iloc[:,3:].copy()
gdp_quarter.columns = ['year_quarter','GDP1','GDP2']
gdp_quarter.index = range(len(gdp_quarter.index))
gdp_quarter
```

```python
_recession_start = []
_recession_end = []
_recession_bottom = []
start_flag,end_flag = 0,0
g_data = gdp_quarter.GDP1.values
d_data = gdp_quarter.year_quarter.values
for i in range(1,len(gdp_quarter.index)):
    if g_data[i] <= g_data[i-1] and start_flag == 0:
        start_flag = 1
    elif g_data[i] <= g_data[i-1] and start_flag == 1:
        start_flag = 2
        _recession_start.append(d_data[i-2])
        low = d_data[i]
    elif g_data[i] <= g_data[i-1] and start_flag == 2:
        low = d_data[i] if d_data[i] <= low else low
    elif g_data[i] > g_data[i-1] and start_flag == 0:
        continue
    elif g_data[i] > g_data[i-1] and start_flag == 1:
        start_flag = 0
    elif g_data[i] > g_data[i-1] and start_flag == 2:
        if end_flag == 0:
            end_flag = 1
        else:
            end_flag = 0
            start_flag = 0
            _recession_end.append(d_data[i-2])
            _recession_bottom.append(low)
#正常状态需要对最后一个值进行特殊吹的，但是嫌麻烦，而且在这个数据下没有必要。
```

```python
_recession_start,_recession_end,_recession_bottom
```

```python
house = city.groupby(["State","RegionName"]).mean()
# 这里看出有重复的地址所以我们取平均来得到数据
house
```

```python
house = house.loc[:,'2000-01':]
house
```

```python
house_quarter = house.iloc[:,:2].copy()  # 得到结构
for i in range(1,int(len(house.columns)/3)+1):
    ye_mo = house.columns[3*i-1]
    name = ye_mo[:4]+'q'+str(int(int(ye_mo[-2:])/3))  #生成名字
    house_quarter = pd.merge(house_quarter,pd.DataFrame(house.iloc[:,3*i-3:3*i].mean(axis = 1),columns= [name]),left_index=True,right_index=True)
    # 将新的到的和原来合并
house_quarter = pd.merge(house_quarter,pd.DataFrame(house.iloc[:,-2:].mean(axis = 1),columns= ['2016q3']),left_index=True,right_index=True)
# 最后剩两个单独处理
house_quarter.drop(['2000-01','2000-02'],axis=1,inplace=True)
house_quarter
```
