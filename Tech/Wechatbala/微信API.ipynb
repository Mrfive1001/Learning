```python
import itchat
itchat.auto_login(hotReload=False)
itchat.run()
itchat.send('Hello, filehelper', toUserName='filehelper')
```

### 1、登陆
短时间关闭程序后重连.这样即使程序关闭，一定时间内重新开启也可以不用重新扫码。最简单的用法就是给auto_login方法传入值为真的hotReload。方法会生成一个静态文件itchat.pkl，用于存储登陆的状态。简单使用这种方式就行。  
itchat.auto_login(hotReload=True)

### 2、回复
itchat提供五种回复方法，建议直接使用send方法。send(msg='Text Message', toUserName=None)  
所需要的值：
* msg：消息内容
* '@fil@文件地址'将会被识别为传送文件，'@img@图片地址'将会被识别为传送图片，'@vid@视频地址'将会被识别为小视频
* toUserName：发送对象，如果留空将会发送给自己，但是有时候发给自己可能会收不到，所以最好发给filehelper

```python
itchat.send('发给你一个图片看咯', toUserName='filehelper')
itchat.send('@img@%s'%'hhh.png', toUserName='filehelper')
```

### 3、账号
#### 好友
主要有get_friends(),search_friends()

```python
my_friends = itchat.get_friends() # 返回完整好友列表list 每个好友一个字典dict
print(my_friends[0])
# 统计男女比例
male = 0
female = 0
for friend in my_friends:
    if friend['Sex'] == 1:
        male += 1
    elif friend['Sex'] == 2:
        female += 1
print('我微信里男生有%d人，女生有%d人'%(male,female))
```

```python
def cal(name):
    chatrooms = itchat.search_chatrooms(name=name)
    if chatrooms:
        chatroom = itchat.update_chatroom(chatrooms[0]['UserName'],detailedMember=True)
        male = 0
        female = 0
        unknow = 0
        for person in chatroom['MemberList']:
            if person['Sex'] == 1:
                male += 1
            elif person['Sex'] == 2:
                female += 1
            else:
                unknow += 1
        print('%s群里男生有%d,女生有%d,未知有%d，男女比为约为%f'%(name,male,female,unknow,male/female))
    else:
        print('没找到相应的群组')
#cal('2017就业实践部')
#cal('北航2017研究生新生官方群')
cal('流利阅读VIP专享-10期25班')
```
