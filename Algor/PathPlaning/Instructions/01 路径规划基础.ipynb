## 2. 路径规划基础
定义：路径规划是指从起始点到目标点规划出一条满足约束或者说不碰到障碍物的一大类算法。利用算法来处理路径规划问题的一般步骤是：
1. 对图像或者得到的数据进行处理
2. 挑选合适的算法得到一条可行的路径
3. 评价路径的优劣(规划时间、规划长度)
4. 运动物体跟踪全局路径
我们主要研究前三个步骤中的一些算法。

## 3. 图像处理
这里的图像处理与计算机视觉中所讲的图像处理不同，是指将原始的图像或者数据处理成算法所需要形式的方法，也叫做地图建模，例如一定形式的二维数据或者某种数据结构。常见的处理方法有许多，这里不一一列举，仅仅介绍两种之后会用到的的算法。
### 栅格法
栅格法是指将原始地图数据处理成等大小的方块，每个方块中的数据表示是否有障碍物，而方块分割的大小是影响后期算法的关键因素。方块较大的话，所需存储信息较少，但是表示原始图像误差较大，而方块较小的时候，环境信息较为准确，但是所需空间较大，且会影响后期算法的维度。
### PRM
PRM全称是随机路径图，表示在原始图像中随机生成N个点构成的图，之后利用这N个点构成的图来表示原始图像，从而能够大大缩小原始图像的维度，为更多的算法应用到路径规划组提供了一种新的可能。PRM生成图的基本步骤是：
1. 在图中生成N个有效点(不在障碍物上的点)
2. 对每个点A，寻找邻域R内的所有点(或者寻找周围最近的m个点)
3. 对找到的每个点B，判断时候AB是否有效(不经过障碍物)
4. 如果有效将B加到A的邻居节点中
5. 得到所有点的邻居节点，得到处理后的图

## 4. 路径规划算法
得到了前期处理的图之后，我们可以利用路径规划算法从起始点到目标点进行规划，规划算法主要可以分为以下几类：
* 图搜索算法
    * A星
    * Dijkstra
    * BFS(Best first search)
* 智能优化算法
    * ACO
    * GA
* 数学概率学算法
    * 人工势场
    * RRT
