# Matplotlib学习记录
可视化是数据科学中很重要的一个步骤，能够为结果和过程提供一个非常好的展示效果，增加可信度。在Python中matolotlib是一个非常强大的图形库。照旧我们先对导入库进行规范：

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1、基本图像
Matplotlib画图主要是有Figure和axes对象。一般使用subplot进行整个图像的分配较为方便。

```python
fig = plt.figure()    #创建一个新的图像
plt.subplot(2,2,1)  #给图像划分区域
x = np.linspace(0,10,200)
plt.plot(x,x**2,x,x)
plt.subplot(2,2,2)
plt.plot(x,x**3)
plt.subplot(2,1,2)
plt.plot(x,np.sin(x))

plt.show()
```

### 2、其他形式图
掌握了基本的画图结构，我们就来看一下各种各样的图像，直方图，散点图等。

```python
X = np.random.randn(1024)
Y = np.random.randn(1024)
plt.scatter(X,Y) #散点图
plt.scatter(X,Y+1)
plt.show()
```

```python
x = np.arange(10)
y = 10*np.random.rand(10)
plt.bar(x,y) #直方图
plt.show()
```

我们基本可以画出我们想要的图像了，但是如果我想对刻度，对图做一些解释说明的话，我们就要使用其他的一些性质。使用set_xticks,set_xticklabels,set_xlabel,set_title,对于Y轴的操作同理。

```python
import matplotlib.pylab as pylab
pylab.mpl.rcParams['font.sans-serif'] = ['SimHei'] #解决不显示中文
pylab.mpl.rcParams['axes.unicode_minus'] = False #解决保存图像是负号'-'显示为方块的问题  
fig = plt.figure()
ax = plt.subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one','two','three','four','five'],rotation = 30)
ax.set_title('zhou')
ax.set_xlabel('次数')
plt.show()
```

我们已经对图像有了相当的控制，接下来我们更进一步加上图例和注释。图例本来意思是给每个线进行解释，从而进行区分。

```python
x = np.linspace(0,10,200)
plt.plot(x,np.sin(x),label = 'sinx')
plt.plot(x,np.cos(x),label = 'cosx')
plt.legend()   #必须加上表示显示这个图例
plt.show()
```

总结：使用Matplotlib的时候要记住他是一个表示x-y关系的展示形式，并没有特别的结构，所以只要想清楚了x-y的映射即可以画出我们想要的图形。
